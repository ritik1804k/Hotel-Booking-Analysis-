# Columns Discription

* Nationality : Player representing that particular nation.
* FIFA Ranking : The teams of the men's member nations of FIFA, football's 
world governing body, are ranked based on their game results with the most successful teams being ranked highest.
*  National Team Kit Sponsor : A particular brand sponsoring that football team and preparing jersey for them.
* Position : The specific role that a player takes on the field is referred to as their "position".
* National Team Jersey Number : Marking of a number behind each player's jersey.
* Player DOB : Player Date of Birth.
* Club : Club are sometimes used interchangeably by supporters, typically referring to the team within the club playing in the highest division or competition.
* Player Name : Name used in Fifa.
* Appearances : A set number of games that the player must appear on the ground for the club.
* Goals Scored : A goal is scored when the whole of the ball passes over the goal line, between the goalposts and under the crossbar, provided that no offence has been committed by the player.
* Assists Provided : Assists are awarded to a player who makes the final pass or cross before a goal is scored.
* Dribbles per 90 : Dribble per 90 means travelling with the ball in between 90 minutes.
* Interceptions per 90 : Number of times stealing the ball from your opposition in between 90 minutes.
* Tackles per 90 : To seize and throw down or stop (an opposing player with the ball) in football.
* Total Duels Won per 90 : The defensive player stopped the progression of the attacking player with the ball and didn't commit a foul, the defensive duel is considered won in those 90 minutes (and the linked offensive duel is considered lost).
* Save Percentage : It is given as by dividing the number of saves by the total number of shots on goal.
* Clean Sheets :  A goalkeeper is said to "keep a clean sheet" if they prevent their opponents scoring any goals during an entire match.
* Brand Sponsor/Brand Used : Brand sponsorship is when a brand acts as a sponsor. The brand would choose the cause to support, be it an event, a small business, or a charity.
